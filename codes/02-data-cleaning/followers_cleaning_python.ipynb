```python
import pandas as pd
```

```python
data=pd.read_csv("../../data/00-raw-data/followers.csv")
data
```

```python
#feature engineering
data[["protected","verfied"]]=data[["protected","verfied"]].replace(False,0)
data[["protected","verfied"]]=data[["protected","verfied"]].replace(True,1)
```

```python
data.drop(columns=(["description","location"]),axis=1,inplace = True)
```

```python
data.to_csv("../../data/01-modified-data/cleaned_followers.csv",index=False)
```
